```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from matplotlib.font_manager import FontProperties
zhfont = FontProperties(fname="C:\Windows\Fonts\SimHei.ttf")
plt.rcParams['font.sans-serif'] = ['SimHei']    # 图例中文显示问题
plt.rcParams['axes.unicode_minus'] = False
sns.set(font = 'SimHei')   # 解决seaborn中文显示问题
```

```python
'''
对图表整体颜色、比例等进行风格设置，包括颜色色板等
调用系统风格进行数据可视化

set() / set_style() / axes_style() / despine() / set_context()
'''
```

```python
# 创建正弦函数及图表
def sinplot(flip =1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i*0.5) * (7-i)*flip)
sinplot()
```

```python
# 用seaborn默认风格
sns.set()
sinplot()
plt.grid(True, linestyle = '--', color = 'white', alpha = 0.9)
```

```python
# set_style()
# 切换seaborn风格
# 风格选择包括“white”,"dark","whitegrid","darkgrid","ticks"
fig = plt.figure(figsize= (14,5))
ax1 = fig.add_subplot(1,2,1)
sns.set_style("darkgrid")
data = np.random.normal(size = (20,6)) + np.arange(6)/2

sns.set(font = 'SimHei')
sns.boxplot(data = data)
plt.title("风格 = whitegrid")

sns.set(font = 'SimHei')
ax2 = fig.add_subplot(1,2,2)
sns.set_style("ticks")
sinplot()
plt.title("没有标题")

# 就很奇怪，seaborn解决中文显示问题，每个图表都得加上 sns.set(font = 'SimHei')，
# 另外就是他的style更改，效果延迟，延迟效果自己测试
```

```python
# despine 坐标轴设置
# sns.despine(
#     fig=None,
#     ax=None,
#     top=True,
#     right=True,
#     left=False,
#     bottom=False,
#     offset=None,
#     trim=False,
# )
sns.set_style("ticks")   # 设置风格

fig = plt.figure(figsize =(18,5))
plt.subplots_adjust()

ax1 = fig.add_subplot(1,3,1)
sinplot()
sns.despine()  # 默认只展示左侧和下方的坐标轴

ax2 = fig.add_subplot(1,3,2)
sns.violinplot(data = data)
sns.despine(offset = 10,  # 坐标轴与偏移量，可以自行测试
            trim = False,  # 当坐标轴偏移量很大时，可以把坐标轴铺长，可以自行测试
           )
# offset 与坐标轴之间的偏移
# trim: 为true时，将坐标轴限制在数据的最大值和最小值范围内

ax3 = fig.add_subplot(1,3,3)
sns.boxplot(data = data, palette="deep")
sns.despine(left =True, right = False)
# left,right,bottom, top ，为True时，坐标轴不显示，为false则显示

# 很蛋疼的一点，对其中一个子图设置时，别的子图也变了
```

```python
# 刚蛋疼完，就有解法了
# axes_style() 设置局部图表风格，可学习和with配合用法
plt.figure(figsize = (14,5))
with sns.axes_style("darkgrid"):
    plt.subplot(121)
    sinplot()
# 设置局部图表风格，用with做代码块区分
sns.set_style('ticks')
plt.subplot(122)
sinplot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']    # 图例中文显示问题
plt.rcParams['axes.unicode_minus'] = False
sns.set(font = 'SimHei')

# 设置风格、尺度
sns.set_style("darkgrid")
sns.set_context("talk")

# 不发出警告
import warnings
warnings.filterwarnings('ignore')
```

## 直方图

```python
# 直方图 distplot()
plt.figure(figsize = (10, 6))
rng = np.random.RandomState(1)
s = pd.Series(rng.randn(100)*100)
sns.distplot(s,   # 数据
            bins = 10,   # 柱子个数
             hist = True,   #是否显示柱子
             kde = False,   # 是否显示密度曲线
             norm_hist= True, # 是否按密度显示
             rug = True,   # 是否显示地毯图
             vertical= False,  # 是否水平显示
             color = 'y',    # 颜色
             label = 'distplot',   # 标签
             axlabel= 'x'   # x轴标签
            )
plt.legend()
```

```python
# 1、直方图 - distplot()
# 颜色详细设置

plt.figure(figsize = (10,6))
sns.distplot(s,rug = True, 
            rug_kws = {'color':'g'} ,   
            # 设置数据频率分布颜色
            kde_kws={"color": "k", "lw": 1, "label": "KDE",'linestyle':'--'},   
            # 设置密度曲线颜色，线宽，标注、线形
            hist_kws={"histtype": "stepfilled", "linewidth": 1,"alpha": 1, "color": "g"})  
            # 设置箱子的风格、线宽、透明度、颜色
            # 风格包括：'bar', 'barstacked', 'step', 'stepfilled'
```

## 密度图

```python
# kdeplot()
# 单个样本数据密度分布图
sns.set_context('notebook')
sns.kdeplot(s,
           shade = True,   # 是否填充
            color ='r',   # 颜色
            vertical=False   # 是否设置水平
           )

sns.kdeplot(s,bw=0.2, label="bw: 0.2",
            linestyle = '-',linewidth = 1.2,alpha = 0.5)
sns.kdeplot(s,bw=2, label="bw: 2",
            linestyle = '-',linewidth = 1.2,alpha = 0.5)
# bw → 控制拟合的程度，类似直方图的箱数

sns.rugplot(s, height=0.1, color = 'k', alpha = 0.8)

```

```python
# 两个样本的密度图
rng = np.random.RandomState(2)
df= pd.DataFrame(rng.randn(100,2),
                columns=['A','B']
                )
plt.figure(figsize =(10, 6))
# 两个维度数据生成曲线密度图，以颜色作为密度衰减显示，可以对照散点图理解
sns.kdeplot(df['A'],df['B'],
           cbar = True,   # 是否显示颜色图例,
            shade = True,  # 是否填充
            cmap = 'Reds',   # 调色盘
            shade_lowest = False,   # 最外围是否填充颜色，
            n_levels = 10   # 曲线个数
           )

sns.rugplot(df['A'], color = 'g', axis = 'x', alpha = 0.5)
sns.rugplot(df['B'], color = 'r', axis = 'y', alpha = 0.5)
plt.scatter(df['A'], df['B'], marker = '.', color = 'b')
```

```python
# 两个样本密度分布图
# 多个密度图
rng1 = np.random.RandomState(2)
rng2 = np.random.RandomState(3)
df1 = pd.DataFrame(rng1.randn(100,2)+2, columns=['A','B'])
df2 = pd.DataFrame(rng2.randn(100,2)-2, columns=['A','B'])

plt.figure(figsize = (10,6))
sns.kdeplot(df1['A'], df1['B'], cmap = 'Greens',
           shade = True, shade_lowest=False,label = 'df1',
            cbar = True
           )
sns.kdeplot(df2['A'], df2['B'], cmap = 'Blues',
           shade = True, shade_lowest= False,label = 'df2',
            cbar = True
           )
```

```python
rs = np.random.RandomState(33)  # 生成一组伪随机数
d = pd.DataFrame(rs.normal(size = (100, 26)))   # 正态分布随机数
d.head()
```

```python
corr = d.corr()   # 求解相关性矩阵表格
corr.head()
```

```python
# 创建数据
mask = np.zeros_like(corr, dtype= np.bool)
mask[np.triu_indices_from(mask)] = True     # 返回方阵的上三角矩阵的索引
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# 设置调色盘
plt.figure(figsize=(10,8))
sns.heatmap(corr,
           mask= mask,
            cmap= cmap,
            vmax = 0.3,
            center= 0,
            square= True,
            linewidths= 0.2
           )
```

```python
# 2、热图 - heatmap()
# 绘制半边热图

sns.set(style="white")
# 设置风格

rs = np.random.RandomState(33)
d = pd.DataFrame(rs.normal(size=(100, 26)))
corr = d.corr()   # 求解相关性矩阵表格
# 创建数据

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# 设置一个“上三角形”蒙版

cmap = sns.diverging_palette(220, 10, as_cmap=True)
# 设置调色盘

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=0.2)
# 生成半边热图
```

```python
# mask参数是设置蒙版用的，如果不设置的话，效果如下
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=0.2)
```
