### 更改值

```python
# 创建一个dataframe
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2],
                 )
df
```

```python
# 指定条件更改值
# 将所有小于0的值改为0
df[df<0] = 0
df
```

```python
# 将第一列大于5的值改为100
df['one'] = df['one'].apply(lambda x: 100 if x>5 else x)
```

```python
# 同样的，也可以用上述方法按照条件新增列
df['new'] = df['three'].apply(lambda x: "10" if x >10 else '10-')
```

### 新增列

```python
# 创建一个dataframe
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2],
                 )
df
```

```python
# 简单新增一列
df['five'] = [1,2,3,4,5,6,7,8]
df
```

```python
# 按照条件新增列，把five分成 1-3, 4-6， 7-8
def rename(x):
    if x >=7 and x <= 8:
        return "7-8"
    elif x >=4 and x <= 6:
        return "4-6"
    elif x >=1 and x <= 3:
        return "1-3"
    else:
        return "qita"
    
df['six'] = df['five'].apply(rename)
df
```

### 排序

```python
# 创建一个dataframe
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2],
                 )
df
```

```python
# 按照 one列降序排列
df.sort_values(by = 'one', ascending= False)
```

```python
# 按照索引1排序，倒序排列
df.sort_index(level= 0, ascending=False)
```

```python
# 按照索引1倒序，索引2正序排列
df.sort_index(level=[0,1], ascending=[False, True])
```

### 添加小计

```python
# 创建一个dataframe
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2]
                 )
df.index.names = ['index1','index2']
df = df.reset_index()
df
```

```python
df2 = df.groupby(['index1','index2']).sum()
df2
```

```python
# 方法1，参考链接 https://stackoverflow.com/questions/41383302/pivot-table-subtotals-in-pandas?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa
# 如果有三个索引，这个方式就不好办了
```

```python
pd.concat([
    d.append(d.sum().rename((k, '小计')))
    for k, d in df2.groupby(level =0)
]).append(df2.sum().rename(("汇总","汇总")))
```

```python
# 方法2，参考链接 https://github.com/chris1610/sidetable/tree/4b1b4e1f34d6991e62144adfaa67a034d01ddfd9
# 支持多个索引，多重小计，这个包还挺好
```

### 表格转成图片

```python
# 方法1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import six

df = pd.DataFrame()
df['date'] = ['2016-04-01', '2016-04-02', '2016-04-03']
df['calories'] = [2200, 2100, 1500]
df['sleep hours'] = [2200, 2100, 1500]
df['gym'] = [True, False, False]

def render_mpl_table(data, col_width=3.0, row_height=0.625, font_size=14,
                     header_color='#40466e', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in six.iteritems(mpl_table._cells):
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax

render_mpl_table(df, header_columns=0, col_width=2.0)

# plt.savefig('123.png')  # 保存图片到当前运行路径
```

### 条件格式

> 参考链接 https://pandas.pydata.org/pandas-docs/stable/user_guide/style.html

```python
# 设置数字颜色
# 创建一个dataframe
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['a','b','c','d','e','f','g','h']
# index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2]
                 )
df
```

```python
def change_color(x):
    if x < 0:
        color = 'green'
    elif x == 0 :
        color ='black'
    else:
        color = 'red'
    return('color:%s' % color)
```

```python
df = df.style.format({"one": "{:,.1F}",
                 'two': '{:.2%}'
                })
df.applymap(change_color)
# 此时df是style对象，不再是dataframe对象
```

```python
# 还可以只对一列设置格式
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['a','b','c','d','e','f','g','h']
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2]
                 )
df
df = df.style.format({"one": "{:,.1F}",
                 'two': '{:.2%}'
                })
df.applymap(change_color, subset = pd.IndexSlice[:,['three']])
```

```python
# 对填充颜色的改变
df.bar(subset = ['four'], color = 'green', vmin = -10, vmax = 15)
```

### 设置表头格式

```python
# 创建一个frame
index1 = ['A'] * 3 + ['B'] * 2 + ['C'] * 3
index2 = ['male'] * 2 + ['female'] * 4 + ['qita'] *2
df = pd.DataFrame(np.random.randn(8, 4) * 10, columns= ['one', 'two', 'three', 'four'],
                  index= [index1, index2]
                 )
df.index.names = ['index1','index2']
df
```

```python
# 将表头颜色改为暗夜蓝
# 不懂这里是css还是html语言，反正只需将对应的参数设置好，直接传入就行了
cell_hover = {
    'selector': 'thead th', 
    'props': [
        ('background-color', '#002060'),
        ('color', 'white'),
        ('caption_side', 'bottom'),
        ('font_size', '12'),   # 字体大小
        ('text-align', 'right')  # 对齐方式
        ]
    
}
df.style.set_table_styles([cell_hover])
```
